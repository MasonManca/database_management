#### Normal Forms:

#### FD = a “functional dependency” 
* Written as: eid → name, dept, dept_name
* Implying: eid → name, eid → dept, and eid → dept_name

### 3 special kinds of FDs
1. Key FDs are of the form X -> Y where X contains a key
  * X is a superkey
  * databse enforces these for us
1. Trivial FDs are of the form X-> Y such that $Y \subseteq X$ (subset)
  * dept -> dept
1. Non-key, Non-Trivial FDs
  * The rest: The non-key FDs that aren't trivial
  * These are 'bad'

### Normalization:
* “Decomposes” a table into smaller tables 
  * Based on current attribute dependencies, follows decomposition algorithm
* Assumes more than 1 relation

### Normal Forms:
1. 1NF = atomic (assumed)
2. 2NF = a relation is 2NF if...
   1. Every non-key attribute is fully dependent on each candidate key
   2. Cannot be partially dependent
3. 3NF = a relation is 3NF if for every non-trivial FD X->Y:
   1. X -> is a key dependency (X superkey)
   2. Y is a part of some candidate key
   3. Only applies if you have a composite key
4. BCNF = subset of 3NF, if every non-trivial FD is a key dependency

## Normal Forms Cont.

### Basic Idea of Normalization:
* Repeatedly decompose a table to remove "bad" FDs
- Generates a binary tree of decomposed relations w/ original table as root
  - Used the bad functional dependency to split relation

To decompose a table using 'bad' FD X-Y: R(WXY)
1. Remove Y from original table $R_1(WX)$, same key
2. Create a table out of X,Y $R_2(XY)$ where X is the key
3. Make X the key $R_2(X^KY)$
4. Repeat for $R_1$ and $R_2$
<br>

- Order matters in the decomposition algorithm

Enrollment(student_id PK,course_id PK, instr_id, student_name, instr_name )
* student_id -> student_name (2)
* course_id -> instr_id (3)
* instr_id -> instr_name ( Not 3NF) (1)
This is 1NF normalization form.
