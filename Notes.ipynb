"Bad" references = a keys which reference another table where the instance does not exist

We can prevent bad references by using "foreign keys"
    * States that column values in one table must come from a key in another table
        ex. v_id must be a number exisiting in an alternate table
    * requires vehicle rows to refer to vehicleType rows ( or to be NULL )

Enforcing foriegn key constraints is known as "referential integrity"

Foreign Key summary:
    * Must reference the entire primary key
        - if R(_x,_y,z), S(_u,x,y)
        then, S.(x,y) references R.(x,y)
    * Must have null or matching fk value
    * enforced by DBms
    * Can be NULL

- DOUBLE

    - TEXT = Character version of a blob

#### Creating Tables SQL

Foreign key format: FOREIGN KEY(att1,att2) REFERENCES branch_name(att1,att2)

CREATE TABLE branch (
    branch_name VARCHAR(50),
    address TINYTEXT NOT NULL,
    phone char(12) NOT NULL,
    PRIMARY KEY (branch_name)
    UNIQUE (address),
    UNIQUE(phone)
);

CREATE TABLE account(
    acct_id INT UNSIGNED,
    acct_name TINYTEXT NOT NULL,
    main_branch VARCHAR(50),
    <!-- Need this data type to be the same as branch_name because it is a FK -->
    PRIMARY KEY (acct_id),
    FOREIGN KEY (main_branch)
    REFERENCES branch(branch_name)
);

    Checkout_date DATE,
